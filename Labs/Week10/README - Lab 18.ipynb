# Simple Hedge Fund Security Selection Challenge

In the following example, we will explore the fundamental principles of portfolio security selection, specifically focusing on the balance between risk and reward. Assume that a hedge fund invests a total of $10,000 split across three different strategies. Each of these strategies has different expected annual returns and different annualized volatility. Here, volatility is a metric that determines the historical liability to change rapidly and unpredictably, especially for the worse. Of course, volatility for investments is generally determined based on historical data that won't necessarily apply to the future, but for this simplified example we will just assume the higher the volatility, the higher the risk, and vice versa. However, the higher the risk, the higher the return, and vice versa. Hence, the objective of the hedge fund is to construct a diversified portfolio aiming to optimize its asset allocation such that it maximizes the returns while controlling for the risk.


Let us assume that the three strategies have the following expected return and volatility.

Strategy 1: Expected Annual Return = 10%, Annualized Volatility = 12%

Strategy 2: Expected Annual Return = 16%, Annualized Volatility = 20%

Strategy 3: Expected Annual Return = 24%, Annualized Volatility = 30%


```python
import numpy as np

# expected annual returns
expected_annual_returns = np.array([0.10, 0.16, 0.24])

# expected annualize volatility
volatilities = np.array([0.12, 0.20, 0.30])
```

Further, let us assume that the correlation coefficients between these strategies are as follows:

Correlation between 1 and 2: 0.6

Correlation between 2 and 3: 0.8

Correlation between 1 and 3: 0.4


```python
# correlation between the strategies
correlation_matrix = np.array([
    [1, 0.6, 0.4],
    [0.6, 1, 0.8],
    [0.4, 0.8, 1]
])
```

The hedge fund allocates, say, 10% in strategy 1 (\$1,000), 20% in strategy 2 (\$2,000), and the remaining 70% in strategy 3 (\$7,000). Then, we can calculate the expected return of the portfolio as follows:

```python
# allocation to each strategy
allocation = np.array([0.10, 0.20, 0.70])

# portfolio return
portfolio_return = np.dot(allocation, expected_annual_returns)

print(f"The expected return on the given portfolio is {portfolio_return * 100}%")
```

This means that the initial investment of \$10,000 yields an expected annual return of \$2,100. But how risky was this investment strategy? Let’s look at the volatility of the portfolio. Given the strategies’ correlation, the volatility of a portfolio can be calculated as follows:

Screenshot 2024-04-07 at 8.23.15 PM.png


Given our numbers, let's calculate the expected annualized volatility for our portfolio:



```python
covariance_matrix = np.outer(volatilities, volatilities) * correlation_matrix

# portfolio volatility
portfolio_volatility = np.sqrt(allocation @ covariance_matrix @ allocation.T)

print(f"The volatility on the given portfolio is {portfolio_volatility * 100:.2f}%")
```

As you can see, since the hedge fund made a heavy investment in a highly volatile strategy i.e. 3, the exposure to risk is significant making the volatility as high as 24.87%. However, if this bet works out well, the fund will also enjoy a 21% return.

To check how well this hedge fund performed over the market, we can check for its alpha value as follows:

Screenshot 2024-04-07 at 8.39.21 PM.png


Let’s assume we are comparing the performance of this hedge fund with S&P 500 index which has historically given a return of 10%, on average. If the risk-free rate of return is 4% and the beta for this portfolio is 1.2, we can calculate the alpha for this fund as follows:

```python
# portfolio beta value
beta = 1.2

# risk free return
risk_free = 0.04

# index return
index_return = 0.1

# calculating alpha
alpha = portfolio_return - (risk_free + beta * (index_return - risk_free))

print(f"The alpha of the given portfolio is {alpha * 100:.2f}%")
```

The positive alpha of 9.8% suggests that the portfolio has outperformed the benchmark index (S&P in this case) by 9.8% after adjusting for risk.
