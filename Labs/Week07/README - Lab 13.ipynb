# Collateralized stablecoins - DAI and MiniDAI

DAI is a sophisticated decentralized stablecoin pegged to the US Dollar, operating on the Ethereum blockchain. Managed by the MakerDAO system, it employs a dynamic system of Collateralized Debt Positions (CDPs), autonomous feedback mechanisms, and incentives for external actors to stabilize its value. Users can mint DAI by locking collateral assets (such as ETH or other ERC-20 tokens) into a smart contract. The system ensures over-collateralization to maintain stability and solvency. The peg to the US Dollar is maintained through external market mechanisms and internal smart contract-enforced rules that adjust parameters like stability fees (interest rates) and collateralization ratios based on market conditions and system health. This model allows DAI to remain relatively stable against the US Dollar, despite the volatility inherent in cryptocurrency markets.

MiniDAI, as outlined in our examples, is a simplified representation of DAI designed for our la. It illustrates the basic mechanisms of a stablecoin by allowing users to mint tokens in exchange for depositing a collateral asset, such as LabUSD, at a predetermined peg rate (e.g., 1 MiniDAI = 1 LabUSD). Users can also burn MiniDAI to withdraw their collateral, aiming to maintain the token's value stability relative to the peg. Although MiniDAI simplifies the complex mechanisms found in real-world stablecoins like DAI, it captures the essence of the collateralized issuance and redemption processes that underpin the operation of stablecoins.

## Stabilizer contract

The Stabilizer smart contract is designed to maintain the peg of a stablecoin, such as MiniDAI, to its intended value (e.g., 1 MiniDAI = 1 LabUSD), by engaging in arbitrage strategies with a decentralized exchange like Uniswap V3. It operates by continuously monitoring the exchange rate of MiniDAI to LabUSD on the market. If the price of MiniDAI deviates from its peg, the Stabilizer intervenes by either minting and selling MiniDAI to decrease its market price when above the peg or purchasing and burning MiniDAI to increase its market price when below the peg. This mechanism helps to correct any price discrepancies and stabilize the value of MiniDAI relative to LabUSD, ensuring the stablecoin's value remains consistent with its peg. The contract leverages smart contract technology to automate these adjustments, enabling real-time stabilization actions based on live market conditions, thereby contributing to the overall resilience and reliability of the stablecoin ecosystem.
